```python
NYSE= pd.DataFrame.from_csv(join(dirData, 'NYSE_companylist.csv'))
```

```python
NYSE.sort(columns=['Sector'])
symbols= list(NYSE.index)[:30]
```

```python
G= nx.Graph()
dist_mat= np.zeros([30,30])
G.add_nodes_from(symbols)
for i, s1 in enumerate(symbols):
    for j, s2 in enumerate(symbols):
        p= np.random.poisson(10)
        dist_mat[i,j]= p
        if np.random.randn(1)>1:
            G.add_edge(s1,s2,weight=p)
            
```

```python

pos=nx.spring_layout(G, k=0.3, iterations=200)
nx.draw_networkx_nodes(G,pos,
                       nodelist=symbols,
                       node_color='r',
                       node_size=500,
                       alpha=0.7)
nx.draw_networkx_edges(G,pos,width=1.0,alpha=0.3)
nx.draw_networkx_labels(G,pos,font_size=10,font_family='sans-serif')
plt.axis('off')
plt.savefig("random_example.png") # save as png
plt.show()
```

```python

# Initialize the heatmap (cmap is the color schema)
fig, ax = plt.subplots()
hm = plt.pcolor(dist_mat, cmap=plt.cm.RdYlBu, alpha=0.8)

# Set size of the figure
fig = plt.gcf()
fig.set_size_inches(10,10)

# Put the major ticks at the middle of each cell
ax.set_yticks(np.arange(dist_mat.shape[0])+0.5, minor=False)
ax.set_xticks(np.arange(dist_mat.shape[1])+0.5, minor=False)
# Custom axes: more natural, table-like format
ax.invert_yaxis()
ax.xaxis.tick_top()

# Set labels
ax.set_xticklabels(labels=symbols)
ax.set_yticklabels(labels=symbols)
# Rotate the labels on x-axis (vertical layout)
plt.xticks(rotation=90) 
# Remove the grid-like lines
ax.grid(False)

# Remove the small ticks from the axis
ax = plt.gca()

for t in ax.xaxis.get_major_ticks(): 
    t.tick1On, t.tick2On = False, False 
for t in ax.yaxis.get_major_ticks(): 
    t.tick1On, t.tick2On = False, False  

# Change font size and add color bar on the side
plt.tick_params(labelsize=14)
plt.colorbar()
```
