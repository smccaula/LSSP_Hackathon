```python
def daychange2bins(x, bins=30, scale=2):
    if x == np.nan:
        return 0
    y= round(x*scale)+round(bins/scale) # ... [-0.75,-0.25) [-0.25,0.25) [0.25,0.75) ... 
    if y > bins:
        return int(bins)
    elif y < 0:
        return int(0)
    else:
        return int(y)
```

```python
def KLdivergence(P, Q):
    """ take 2 numpy.ndarray with same length """
    if len(P) != len(Q):
        print('not equal length')
        return None
    elif type(P)!=np.ndarray or type(Q)!=np.ndarray:
        print('not numpy.ndarray')
        return None
    else:
        KL= 0
        for i in list(range(len(P))):
            if Q[i]==0:
                Qi= 2**(-1000)
            else:
                Qi= Q[i]
            if P[i]==0:
                Pi= 2**(-1000)
            else:
                Pi= P[i]
            try:
                KL += Pi*math.log(Pi/Qi,2)
                #print(i, KL)
            except:
                print('@@@@ something is wrong', P[i], Q[i])
        return KL
```

```python
def JSDistance(P, Q):
    hPQ= (P+Q)/2
    return (KLdivergence(P, hPQ)+KLdivergence(Q, hPQ))/2
```

```python
def csv2prob(csv):
    """ read csv file and return numpy.ndarray of probability """
    df= pd.DataFrame.from_csv(csv)
    print('read', csv, 'with length', len(df))
    df['daychange']= float(0)
    for i in list(range(len(df))):
        dc= 100*df['diff'][df.index[i]]/df['close'][df.index[i-1]]
        df['daychange'][df.index[i]]= dc
    df['bin']= df['daychange'].map(daychange2bins)
    dt= {} 
    counts= df['daychange'].count()
    for i in list(range(31)):
        dt[i]= {}
        dt[i]['count']= df['daychange'][df['bin']==i].count()
        dt[i]['min']= df['daychange'][df['bin']==i].min()
        dt[i]['max']= df['daychange'][df['bin']==i].max()
        prob= df['daychange'][df['bin']==i].count()/counts
        dt[i]['prob']= prob
        #print(i, prob)
        dt[i]['surprise']= math.log(1/prob,2)
    arr= np.array([dt[i]['prob'] for i in dt])
    #print(arr)
    return arr
```

```python
def list2prob(the_list):
    for s in list(range(len(the_list))):
        try:
            globals()[the_list[s]]= csv2prob(the_list[s]+'.csv')
        except:
            print('@@@@ csv2prob failed at', the_list[s])   
            
def prob2JSD(the_list):
    dt= {}
    for p in the_list:
        dt[p]= {}
        for q in the_list:
            dt[p][q]= None
    for p in the_list:
        for q in the_list:
            if p != q:
                try:
                    jsd= JSDistance(eval(p), eval(q))
                    dt[p][q]= jsd
                    dt[q][p]= jsd
                    #print(p, q, dt[p][q])
                except:
                    print('@@@@ JSDistance failed at', p, q)
    df= pd.DataFrame.from_dict(dt)
    return df
```
